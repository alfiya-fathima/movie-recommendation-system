```python
import pandas as pd

movies = pd.read_csv("tmdb_5000_movies.csv")
movies.head()
```

```python
movies.columns
```

```python
movies = movies[['original_title','genres','keywords','overview']]
movies.head()
```

```python
movies = movies.dropna()
```

```python
[{'id': 28, 'name': 'Action'}, {'id': 12, 'name': 'Adventure'}]
```

```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return " ".join(L)

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['combined'] = movies['genres'] + " " + movies['keywords'] + " " + movies['overview']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
matrix = cv.fit_transform(movies['combined'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(matrix)
```

```python
def recommend(movie):
    movie = movie.lower()
    
    if movie not in movies['original_title'].str.lower().values:
        print("Movie not found")
        return
    
    index = movies[movies['original_title'].str.lower() == movie].index[0]
    
    distances = list(enumerate(similarity[index]))
    movies_list = sorted(distances, key=lambda x: x[1], reverse=True)[1:6]
    
    for i in movies_list:
        print(movies.iloc[i[0]].original_title)
```

```python
recommend("Avatar")
```
